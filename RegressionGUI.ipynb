```python

import tkinter as tk 
import pandas as pd
import statsmodels.api as sm
from tkinter import filedialog
from tkinter import *
```

```python
def getCSV():
    
    global df
    
    import_file_path = filedialog.askopenfilename()
    df = pd.read_csv(import_file_path)
    
    
    
def getSavedInfo():
    
    global DV
    global IV1
    global IV2
    
    DV = entry1.get()
    IV1 = entry2.get()
    IV2 = entry3.get()
    
def splitData():
    
    global yin
    global Xin
    yin = df.loc[:,DV]
    col_start = df.columns.get_loc(IV1)
    col_end = df.columns.get_loc(IV2)
    col_end = col_end+1
    Xin = df.iloc[:,col_start:col_end]
    
    
def olsRegression():
    splitData()
    Xnew = sm.add_constant(Xin)
    LinearModel = sm.OLS(yin,Xnew).fit()
    predictions = LinearModel.predict(Xnew)
    Output1 = (LinearModel.summary2().tables[0])
    Output2 = (LinearModel.summary2().tables[1])
    Output3 = (LinearModel.summary2().tables[2])
    global df1 
    df1 = pd.DataFrame(Output1)
    global df2
    df2 = pd.DataFrame(Output2)
    global df3 
    df3 = pd.DataFrame(Output3)  

def dftoExcel():
    
    filename = entry4.get()
    filename = filename + '.xlsx'
    with pd.ExcelWriter(filename) as writer:
        
        df.to_excel(writer,sheet_name='Input')
        df1.to_excel(writer,sheet_name='OLS Regression result')
        df2.to_excel(writer,sheet_name='OLS Regression result1')
        df3.to_excel(writer,sheet_name='OLS Regression result2')
        

    
def opennewwindow():
    
    root = tk.Tk()
    root.geometry("800x500")
    
    global entry1
    global entry2
    global entry3
    global entry4
    
    tk.Button(root,text="Import CSV",width = 20,command=getCSV).grid(column = 0, row = 0,pady=10, sticky = W)
    
    tk.Label(root,text="Please, enter data: ").grid(column=0, row=1,pady= 10, sticky= W)
    
    tk.Label(root,text="Dependent Variable: ").grid(column=0, row=2, pady = 5, sticky = W )
    entry1 = tk.Entry(root,width = 50)
    entry1.grid(column=1, row=2, pady=5, sticky=W)
    
    tk.Label(root,text="Independent Variable1: ").grid(column=0, row=3, pady = 5, sticky = W )
    entry2 = tk.Entry(root,width = 50)
    entry2.grid(column=1, row=3, pady=5, sticky=W)

    tk.Label(root,text="Independent Variable2: ").grid(column=0, row=4, pady = 5, sticky = W )
    entry3 = tk.Entry(root,width = 50)
    entry3.grid(column=1, row=4, pady=5, sticky=W)
    
    tk.Button(root, text="Save", width = 20, command=getSavedInfo).grid(column=0, row=5, pady = 10, sticky =W)
    
    tk.Button(root, text="Generate Results", width = 20, command=olsRegression).grid(column=0, row=6, pady = 10, sticky =W)
    
    tk.Label(root,text="Name of excel file: ").grid(column=0, row=7, pady = 5, sticky = W )
    entry4 = tk.Entry(root,width = 50)
    entry4.grid(column=1, row=7, pady=5, sticky=W)
    
    tk.Button(root, text="OLS Regression Result", width = 20, command = dftoExcel).grid(column = 0, row=8, pady = 5, sticky = W)
    
    root.mainloop()

opennewwindow()





```
